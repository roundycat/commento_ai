데이터 전처리

```python
import cv2
import numpy as np
import os

image_files = [
    "1000911.jpg", "1000351.jpg", "1001113.jpg", "1000431.jpg",
    "1003796.jpg", "1004499.jpg", "1005066.jpg", "1006121.jpg",
    "1047447.jpg", "1011469.jpg"
]

output_dir = "preprocessed_images"
os.makedirs(output_dir, exist_ok=True)

for file in image_files:
    img = cv2.imread(file)
    if img is None:
        print(f"{file} 파일을 읽지 못했습니다.")
        continue
    # 크기 조정
    img = cv2.resize(img, (224, 224))
    # 그레이스케일 변환
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # 정규화 및 블러
    norm = gray / 255.0
    blur = cv2.GaussianBlur(norm, (5, 5), 0)
    img_uint8 = (blur * 255).astype(np.uint8)
    # 원본 저장
    base = os.path.splitext(file)[0]
    cv2.imwrite(f"{output_dir}/{base}_proc.png", img_uint8)
    # 좌우 반전 저장
    cv2.imwrite(f"{output_dir}/{base}_flip.png", cv2.flip(img_uint8, 1))
    # 회전
    h, w = img_uint8.shape
    for angle in [15, -15]:
        M = cv2.getRotationMatrix2D((w / 2, h / 2), angle, 1)
        rot = cv2.warpAffine(img_uint8, M, (w, h))
        cv2.imwrite(f"{output_dir}/{base}_rot{angle}.png", rot)
    # 밝기 증강
    brighter = cv2.convertScaleAbs(img_uint8, alpha=1.2, beta=10)
    darker = cv2.convertScaleAbs(img_uint8, alpha=0.8, beta=-10)
    cv2.imwrite(f"{output_dir}/{base}_bright.png", brighter)
    cv2.imwrite(f"{output_dir}/{base}_dark.png", darker)
```
